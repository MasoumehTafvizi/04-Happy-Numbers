- for loop on n characters( after converting to string)
- power of 2
- sum of the powers
- if sum is 1, return True
- check if sum is not in a set of seen numbers

```python
n = 49
def is_happy(n):
    seen_numbers = set()
    while (n != 1) and (n not in seen_numbers):
        seen_numbers.add(n)
        n = sum(int(i) ** 2 for i in str(n))
    return n == 1

is_happy(n)
```
